```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
a_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Finland'],
    'Population (M)': [82.8, 67.2, 11.4, 5.5],
    'Capital': ['Berlin', 'Paris', 'Brussels', 'Helsinki']
})
a_df
```

```python
b_df = pd.DataFrame({
    'Country': ['Germany', 'France', 'Belgium', 'Canada'],
    'HDI': [0.936, 0.901, 0.916, 0.926]
})
b_df
```

```python
inner_merged_df = a_df.merge(b_df, on ="Country")
inner_merged_df 
```

```python
left_merged_df = a_df.merge(b_df, on ="Country", how="left")
left_merged_df
```

```python
right_merged_df = a_df.merge(b_df, on ="Country", how="right")
right_merged_df
```

```python
outer_merged_df = a_df.merge(b_df, on='Country', how='outer')
outer_merged_df
```

```python
countries_df =  pd.read_csv('dictionary.csv')
countries_df.head(3)
```

```python
summer_df = pd.read_csv('summer.csv')
summer_df.head(10)
```

```python
summer_df = summer_df.rename(index=str, columns={"Country": "Code"})
summer_df.head(3)
```

```python
winter_df = pd.read_csv('winter.csv')
winter_df.rename(columns={"Country": "Code"}, inplace=True)
winter_df.head()
```

```python
summer_countries_df = countries_df.merge(summer_df, on ="Code", how="right")
summer_countries_df.head(3)
```

```python
summer_countries_df['Season'] = str("Summer")
summer_countries_df
```

```python
winter_countries_df = countries_df.merge(winter_df, on ="Code", how="right")
winter_countries_df['Season'] = str("Winter")
winter_countries_df.head(3)
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], sort=False)
all_df.tail()
```

```python
all_df[(all_df['Year'] >= 1984 ) ]\
.groupby(['Code']).count().sort_values(by=['Medal'],ascending=False)
```
